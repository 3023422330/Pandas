```python
one = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'], 'A': ['A0', 'A1','A2','A3'], 'B': ['B0', 'B1', 'B2', 'B3']})
```

```python
two = pd.DataFrame({'key': ['K0', 'K1', 'K4', 'K5'], 'C': ['C0', 'C1','C2','C3'], 'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
one
```

```python
two
```

```python
pd.merge(one,two, how='outer') # outer means to collect all data from one and two
```

```python
pd.merge(one,two, how='inner')   # inner means to collect the data where the same row from one and two
```

```python
pd.merge(one,two, how='left') # left means to collect data from one
```

```python
pd.merge(one, two, how='right') # right means to collect data from two
```

```python
one = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'], 'A': ['A0', 'A1','A2','A3'], 'B': ['B0', 'B1', 'B2', 'B3']})
two = pd.DataFrame({'key': ['K0', 'K1', 'K4', 'K5'], 'A': ['A10', 'A11','A12','A13'], 'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
one
```

```python
two
```

```python
pd.merge(one, two, on='key')
```

```python
pd.merge(one, two, on='key', suffixes=('_one', '_two'))
```
