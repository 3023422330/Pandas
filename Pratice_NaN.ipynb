```python
import pandas as pd
```

```python
studf = pd.read_excel("student_excel.xlsx",skiprows=2)
```

```python
studf
```

```python
studf.isnull() 
```

```python
 studf['分数'].isnull()
```

```python
 studf['分数'].notnull()
```

```python
studf.loc[studf['分数'].notnull(),:]   # search data of 分數 is not empty(notnull)
```

```python
studf.dropna(axis='columns', how='all', inplace=True)  # remove the column with all NaN
```

```python
studf
```

```python
studf.dropna(axis='index', how='all', inplace=True) # remove the row with all NaN
studf
```

```python
studf.fillna({'分数':0})
```

```python
studf.loc[:, '分数'] = studf['分数'].fillna(0)
studf
```

```python
# forward fill the NaN in 姓名, use .fillna(method='ffill')
studf.loc[:, '姓名'] = studf['姓名'].fillna(method='ffill')
studf
```

```python
# save excel
# delet the first column with numbers 0,1,2,3,4,... , use index=False
studf.to_excel('student_excel_clean.xlsx', index=False)
```
