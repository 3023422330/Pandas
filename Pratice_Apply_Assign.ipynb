```python
import pandas as pd
```

```python
df = pd.read_csv('beijing_tianqi_2018.csv')
```

```python
df.head()
```

```python
df.loc[:, 'bWendu'] = df["bWendu"].str.replace("℃", "").astype('int32')
```

```python
df.loc[:, 'yWendu'] = df["yWendu"].str.replace('℃', '').astype('int32')
```

```python
df.head()
```

```python
df.loc[:, "wencha"] = df["bWendu"] - df["yWendu"]
df.head()
```

```python
def get_wendu_type(x):
    if x['bWendu'] >33:
        return 'High'
    if x['yWendu'] < -10:
        return 'low'
    return 'Normal'
df.loc[:, 'wendu_type'] = df.apply(get_wendu_type, axis=1)
```

```python
df['wendu_type'].value_counts()
```

```python
df.head(10) 
```

```python
df.assign(yWendu_huashi = lambda x : x['yWendu'] * 9/5+32, 
          bWendu_huashi = lambda x : x['yWendu'] * 9/5+32)
```

```python
df['wencha_type'] = ''
df.loc[df['bWendu'] - df['yWendu']>10, 'wencha_type'] = 'High'
df.loc[df['bWendu'] - df['yWendu']<=10, 'wencha_type'] = 'Normal'
```

```python
def get_tianqi_comment(i):
    if i['tianqi'] == '阴':
        return 'I do not like it'
    if i['tianqi']== "晴~多云" or '阴~多云':
        return 'Ummmm, not good.'
    if i['tianqi'] == '多云':
        return 'so bad'
    if i['tianqi'] == '晴':
        return 'I like it'
df.loc[:, 'tianqi_comment'] = df.apply(get_tianqi_comment, axis=1)
```
