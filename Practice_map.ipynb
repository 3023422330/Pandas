```python
import pandas as pd
stocks = pd.read_excel('互联网公司股票.xlsx')
stocks.head()
```

```python
dict_company_names = {
    'bidu': '百度',
    'baba':'阿里巴巴',
    'iq':  '愛奇藝',
    'jd': '京東'
}
```

```python
stocks['Chinese_company'] = stocks['公司'].str.lower().map(dict_company_names)
```

```python
stocks.head()
```

```python
stocks['CompanyChinese'] = stocks['公司'].map(lambda x: dict_company_names[x.lower()])
stocks.head()
```

```python
stocks['COmpnay_Chinese'] = stocks['公司'].apply(lambda x : dict_company_names[x.lower()])
stocks.head()
```

```python
sub_df = stocks[['收盘', '开盘','高','低','交易量']]
sub_df
```

```python
sub_df.applymap(lambda x : int(x))    # applymap is to change condition, here is to change the float to int
```

```python
stocks.loc[:, ['收盘', '开盘','高','低','交易量']] = sub_df.applymap(lambda x : int(x))
stocks.head()
```
